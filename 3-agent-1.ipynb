## ğŸ”§ 2. Tool ì •ì˜: ëª¨ë“  í•¨ìˆ˜ì™€ ìŠ¤í‚¤ë§ˆ í†µí•©
ì‹¤ìŠµì— ì‚¬ìš©í•  ëª¨ë“  í•¨ìˆ˜(Tool)ì™€ ê·¸ í•¨ìˆ˜ì˜ ëª…ì„¸(JSON Schema)ë¥¼ ì´ ì„¹ì…˜ì—ì„œ í•œ ë²ˆì— ì •ì˜í•©ë‹ˆë‹¤.

```python
# === Tool 1: í™˜ìœ¨ ì¡°íšŒ í•¨ìˆ˜ ë° ìŠ¤í‚¤ë§ˆ ===
def get_exchange_rate(currency: str) -> str:
    """ì§€ì •ëœ í†µí™”(currency)ì™€ ëŒ€í•œë¯¼êµ­ ì›(KRW) ì‚¬ì´ì˜ í˜„ì¬ í™˜ìœ¨ì„ ê°€ì ¸ì˜µë‹ˆë‹¤."""
    if not isinstance(currency, str) or len(currency) != 3:
        return json.dumps({"error": "ìœ íš¨í•˜ì§€ ì•Šì€ í†µí™” ì½”ë“œì…ë‹ˆë‹¤. 3ìë¦¬ ISO ì½”ë“œë¥¼ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤."})

    url = f"https://api.exchangerate-api.com/v4/latest/{currency.upper()}"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        krw_rate = data.get("rates", {}).get("KRW")
        if krw_rate:
            return json.dumps({"rate": krw_rate})
        else:
            return json.dumps({"error": f"'{currency}'ì— ëŒ€í•œ KRW í™˜ìœ¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."})
    except Exception as e:
        return json.dumps({"error": f"API ìš”ì²­ ì¤‘ ì—ëŸ¬ ë°œìƒ: {e}"})

exchange_tool_schema = {
    "type": "function",
    "function": {
        "name": "get_exchange_rate",
        "description": "íŠ¹ì • í†µí™”ì™€ ëŒ€í•œë¯¼êµ­ ì›(KRW) ì‚¬ì´ì˜ ì‹¤ì‹œê°„ í™˜ìœ¨ì„ ì¡°íšŒí•©ë‹ˆë‹¤.",
        "parameters": {
            "type": "object",
            "properties": {"currency": {"type": "string", "description": "í™˜ìœ¨ì„ ì¡°íšŒí•  ê¸°ì¤€ í†µí™”ì˜ 3ìë¦¬ ISO 4217 ì½”ë“œ (ì˜ˆ: USD, JPY)"}},
            "required": ["currency"]
        }
    }
}

# === Tool 2: ìœ„í‚¤ë°±ê³¼ ê²€ìƒ‰ í•¨ìˆ˜ ë° ìŠ¤í‚¤ë§ˆ ===
def search_wikipedia(query: str) -> str:
    """ìœ„í‚¤ë°±ê³¼ì—ì„œ ì •ë³´ë¥¼ ê²€ìƒ‰í•˜ì—¬ ìš”ì•½í•©ë‹ˆë‹¤."""
    try:
        wikipedia.set_lang("ko")
        return wikipedia.summary(query, sentences=2)
    except Exception as e:
        return f"ìœ„í‚¤ë°±ê³¼ ê²€ìƒ‰ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}"

wiki_tool_schema = {
    "type": "function",
    "function": {
        "name": "search_wikipedia",
        "description": "ìœ„í‚¤ë°±ê³¼(Wikipedia)ì—ì„œ íŠ¹ì • ì£¼ì œì— ëŒ€í•œ ì •ë³´ë¥¼ ê²€ìƒ‰í•˜ê³  ìš”ì•½í•©ë‹ˆë‹¤.",
        "parameters": {
            "type": "object",
            "properties": {"query": {"type": "string", "description": "ìœ„í‚¤ë°±ê³¼ì—ì„œ ê²€ìƒ‰í•  í‚¤ì›Œë“œ ë˜ëŠ” ì§ˆë¬¸"}},
            "required": ["query"]
        }
    }
}

# === ì‚¬ìš© ê°€ëŠ¥í•œ í•¨ìˆ˜ë¥¼ ë§¤í•‘ ===
available_functions = {
    "get_exchange_rate": get_exchange_rate,
    "search_wikipedia": search_wikipedia
}
```

## ğŸ” 3. Agent Full-Loop ìˆ˜ë™ êµ¬í˜„ (â­í•µì‹¬â­)
Agentê°€ ì‹¤ì œë¡œ ë™ì‘í•˜ëŠ” ì „ ê³¼ì •ì„ ë‹¨ê³„ë³„ë¡œ ë‚˜ëˆ„ì–´ ì§ì ‘ ì½”ë“œë¥¼ ì‹¤í–‰í•´ë´…ë‹ˆë‹¤.

```python
messages = [{"role": "user", "content": "100ë‹¬ëŸ¬ëŠ” ì§€ê¸ˆ ì–¼ë§ˆì¸ê°€ìš”?"}]
tools = [exchange_tool_schema]

print("ğŸ’¬ 1ë‹¨ê³„: LLMì— Tool í˜¸ì¶œ ìš”ì²­")
response = client.chat.completions.create(model=MODEL, messages=messages, tools=tools, tool_choice="auto")
response_message = response.choices[0].message
messages.append(response_message)

tool_calls = response_message.tool_calls
if tool_calls:
    print(" LLMì´ Tool ì‚¬ìš©ì„ ê²°ì •í–ˆìŠµë‹ˆë‹¤.")
    print("\n 2ë‹¨ê³„: ê²°ì •ëœ Toolì„ ì‹¤ì œë¡œ ì‹¤í–‰")
    for tool_call in tool_calls:
        function_name = tool_call.function.name
        function_to_call = available_functions[function_name]
        function_args = json.loads(tool_call.function.arguments)
        print(f"- ì‹¤í–‰ í•¨ìˆ˜: {function_name}({', '.join([f'{k}={v}' for k, v in function_args.items()])})")

        function_response = function_to_call(**function_args)
        print(f"- ì‹¤í–‰ ê²°ê³¼: {function_response}")

        messages.append({"tool_call_id": tool_call.id, "role": "tool", "name": function_name, "content": function_response})

    print("\n 3ë‹¨ê³„: Tool ì‹¤í–‰ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ìµœì¢… ë‹µë³€ ìƒì„±")
    final_response = client.chat.completions.create(model=MODEL, messages=messages)
    print("\n===== Agentì˜ ìµœì¢… ë‹µë³€ =====")
    print(final_response.choices[0].message.content)
else:
    print(" LLMì´ Tool ì‚¬ìš©ì„ ê²°ì •í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
```

## ğŸ¤¹â€â™‚ï¸ 5. ì‹¬í™”: Multi-Tool í™˜ê²½ì—ì„œì˜ ì„ íƒ (Parallel Function Calling)
LLMì—ê²Œ ì—¬ëŸ¬ ê°œì˜ ë„êµ¬ë¥¼ ì œê³µí•˜ê³ , ìƒí™©ì— ë§ëŠ” ë„êµ¬ë¥¼ ë™ì‹œì— ì„ íƒí•˜ê²Œ ë§Œë“¤ì–´ ë´…ë‹ˆë‹¤.

```python
# 5. ì‹¬í™”: Multi-Tool í™˜ê²½ì—ì„œì˜ ì„ íƒ (ìˆ˜ì •ëœ ì½”ë“œ)

multi_tools = [exchange_tool_schema, wiki_tool_schema]
prompt = "OpenAIëŠ” ì–´ë–¤ íšŒì‚¬ì¸ê°€ìš”? ê·¸ë¦¬ê³  í˜„ì¬ 1ìœ ë¡œ(EUR)ëŠ” ì–¼ë§ˆì¸ì§€ë„ ì•Œë ¤ì£¼ì„¸ìš”."

# ëŒ€í™” ê¸°ë¡ ì´ˆê¸°í™”
messages = [{"role": "user", "content": prompt}]

# --- 1ë‹¨ê³„: LLMì— Tool í˜¸ì¶œ ìš”ì²­ ---
print("===== 1ë‹¨ê³„: Multi-Tool í™˜ê²½ì—ì„œ LLMì˜ Tool í˜¸ì¶œ ê²°ì • =====")
response_1 = client.chat.completions.create(
    model=MODEL,
    messages=messages,
    tools=multi_tools,
    tool_choice="auto"
)
response_message = response_1.choices[0].message

# LLMì˜ tool_calls ê²°ì •ì„ ì¶œë ¥
print("LLMì˜ Tool ì„ íƒ:")
print(response_message.tool_calls)

# ëŒ€í™” ê¸°ë¡ì— LLMì˜ ì‘ë‹µ ì¶”ê°€
messages.append(response_message)

# --- 2ë‹¨ê³„: ê²°ì •ëœ ëª¨ë“  Toolì„ ì‹¤ì œë¡œ ì‹¤í–‰ ---
tool_calls = response_message.tool_calls
if tool_calls:
    print("\n===== 2ë‹¨ê³„: ê²°ì •ëœ ëª¨ë“  Tool ì‹¤í–‰ =====")
    # ë£¨í”„ë¥¼ ëŒë©° ëª¨ë¸ì´ ìš”ì²­í•œ ëª¨ë“  í•¨ìˆ˜ë¥¼ ì‹¤í–‰
    for tool_call in tool_calls:
        function_name = tool_call.function.name
        function_to_call = available_functions[function_name]
        function_args = json.loads(tool_call.function.arguments)

        print(f"- ì‹¤í–‰ í•¨ìˆ˜: {function_name}")

        # í•¨ìˆ˜ ì‹¤í–‰
        function_response = function_to_call(**function_args)

        # í•¨ìˆ˜ ì‹¤í–‰ ê²°ê³¼ë¥¼ ëŒ€í™” ê¸°ë¡ì— ì¶”ê°€
        messages.append(
            {
                "tool_call_id": tool_call.id,
                "role": "tool",
                "name": function_name,
                "content": function_response,
            }
        )

# --- 3ë‹¨ê³„: Tool ì‹¤í–‰ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ìµœì¢… ë‹µë³€ ìƒì„± ---
print("\n===== 3ë‹¨ê³„: Tool ì‹¤í–‰ ê²°ê³¼ë¥¼ ì¢…í•©í•˜ì—¬ ìµœì¢… ë‹µë³€ ìƒì„± =====")
final_response = client.chat.completions.create(
    model=MODEL,
    messages=messages
)

# ìµœì¢… ë‹µë³€ì˜ content ì¶œë ¥
print("LLMì˜ ìµœì¢… ë‹µë³€:")
print(final_response.choices[0].message.content)
```

## ğŸ¤¯ 6. ì‹¬í™”: Agentì˜ ì‹¤íŒ¨ ì¼€ì´ìŠ¤ (Hallucination)
**Agentê°€ ìˆë‹¤ê³  í•­ìƒ ë˜‘ë˜‘í•œ ê±´ ì•„ë‹™ë‹ˆë‹¤.** êµë¬˜í•œ ì§ˆë¬¸ì—ëŠ” ì—‰ëš±í•œ ë„êµ¬ë¥¼ ì„ íƒí•˜ëŠ” 'í™˜ê°(Hallucination)' í˜„ìƒì„ ë³´ì¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# 6. ì‹¬í™”: Agentì˜ ì‹¤íŒ¨ ì¼€ì´ìŠ¤ (Hallucination) - ìˆ˜ì •ëœ ì½”ë“œ

# 'ì›'ì´ë¼ëŠ” ë‹¨ì–´ì™€ 'ê°€ì¹˜'ë¼ëŠ” ë§¥ë½ìœ¼ë¡œ í™˜ê°ì„ ìœ ë„í•˜ëŠ” ì§ˆë¬¸
prompt_hallucinate = "ë¹„íŠ¸ì½”ì¸ ì°½ì‹œì ì‚¬í† ì‹œ ë‚˜ì¹´ëª¨í† ì˜ ì¬ì‚°ì€ ëŒ€í•œë¯¼êµ­ 'ì›'ìœ¼ë¡œ ì–¼ë§ˆì¸ê°€ìš”?"

# ëŒ€í™” ê¸°ë¡ ì´ˆê¸°í™”
messages = [{"role": "user", "content": prompt_hallucinate}]
tools = [exchange_tool_schema, wiki_tool_schema]

# --- 1ë‹¨ê³„: LLMì— Tool í˜¸ì¶œ ìš”ì²­ ---
print(f"===== ì§ˆë¬¸: '{prompt_hallucinate}' =====")
response = client.chat.completions.create(
    model=MODEL,
    messages=messages,
    tools=tools,
    tool_choice="auto"
)
response_message = response.choices[0].message
messages.append(response_message)

tool_calls = response_message.tool_calls
if tool_calls:
    print("\n===== 1ë‹¨ê³„ ê²°ê³¼: LLMì˜ Tool ì„ íƒ =====")
    print("LLMì´ ì˜ëª»ëœ Toolì„ ì„ íƒí–ˆìŠµë‹ˆë‹¤ (Hallucination ë°œìƒ!):")
    print(tool_calls)

    # --- 2ë‹¨ê³„: ì˜ëª» ì„ íƒëœ Toolì„ ì‹¤ì œë¡œ ì‹¤í–‰ ---
    print("\n===== 2ë‹¨ê³„: ì˜ëª» ì„ íƒëœ Tool ì‹¤í–‰ =====")
    for tool_call in tool_calls:
        function_name = tool_call.function.name
        function_to_call = available_functions[function_name]
        function_args = json.loads(tool_call.function.arguments)

        print(f"- ì‹¤í–‰ í•¨ìˆ˜: {function_name}")

        # í•¨ìˆ˜ ì‹¤í–‰ (ì¡´ì¬í•˜ì§€ ì•ŠëŠ” í†µí™”ì´ë¯€ë¡œ ì—ëŸ¬ê°€ ë‹´ê¸´ ê²°ê³¼ë¥¼ ë°˜í™˜)
        function_response = function_to_call(**function_args)
        print(f"- ì‹¤í–‰ ê²°ê³¼ (ì—ëŸ¬): {function_response}")

        # í•¨ìˆ˜ ì‹¤í–‰ ê²°ê³¼ë¥¼ ëŒ€í™” ê¸°ë¡ì— ì¶”ê°€
        messages.append(
            {
                "tool_call_id": tool_call.id,
                "role": "tool",
                "name": function_name,
                "content": function_response,
            }
        )

    # --- 3ë‹¨ê³„: ì˜ëª»ëœ Tool ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ìµœì¢… ë‹µë³€ ìƒì„± ---
    print("\n===== 3ë‹¨ê³„: ì˜ëª»ëœ Tool ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ìƒì„±ëœ ìµœì¢… ë‹µë³€ =====")
    final_response = client.chat.completions.create(
        model=MODEL,
        messages=messages
    )
    print(final_response.choices[0].message.content)

else:
    print("\n(ì´ë²ˆì—ëŠ” Hallucinationì´ ë°œìƒí•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. ëª¨ë¸ì´ ì˜¬ë°”ë¥´ê²Œ íŒë‹¨í•˜ì—¬ Toolì„ ì‚¬ìš©í•˜ì§€ ì•Šê±°ë‚˜, ì˜¬ë°”ë¥¸ Toolì„ ì„ íƒí–ˆìŠµë‹ˆë‹¤.)")
    print("LLMì˜ ë‹µë³€:")
    print(response_message.content)
```

ì´ë²ˆ ê²°ê³¼ëŠ” Agentì˜ ë™ì‘ ë°©ì‹ì— ëŒ€í•´ ë§¤ìš° í¥ë¯¸ë¡œìš´ ì ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.

Hallucination ë°œìƒ: ì˜ˆìƒëŒ€ë¡œ, ëª¨ë¸ì€ 'ì¬ì‚°', 'ì›'ì´ë¼ëŠ” ë‹¨ì–´ì— ë°˜ì‘í•˜ì—¬ get_exchange_rate Toolì„ í˜¸ì¶œí–ˆìŠµë‹ˆë‹¤. ì‹¬ì§€ì–´ ì§ˆë¬¸ì— ì—†ë˜ 'USD'ë¥¼ ìŠ¤ìŠ¤ë¡œ ê°€ì •í•˜ì—¬ í˜¸ì¶œí•œ ê²ƒì€, ì¼ì¢…ì˜ **'ì¶”ë¡ ì  í™˜ê°(Inferential Hallucination)'**ìœ¼ë¡œ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ê¹Œì§€ë§Œ ë³´ë©´ ìš°ë¦¬ì˜ ì˜ë„ëŒ€ë¡œ ëª¨ë¸ì´ ì‹¤íŒ¨í•œ ê²ƒì²˜ëŸ¼ ë³´ì…ë‹ˆë‹¤.
